```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('/content/Salaries.csv')
df.head()
```

```python
# Basic Data Exploration
print("Basic Data Exploration:")

# Number of rows and columns
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])

# Data types of each column
print("\nData types:\n",df.dtypes)

# Check for missing values in each column
print("\nMissing values:\n",df.isnull().sum())
```

```python
# Descriptive Statistics
print("\nDescriptive Statistics:")
print("Mean:", df['TotalPay'].mean())
print("Median:", df['TotalPay'].median())
print("Mode:", df['TotalPay'].mode())
print("Minimum:", df['TotalPay'].min())
print("Maximum:", df['TotalPay'].max())
print("Range:", df['TotalPay'].max() - df['TotalPay'].min())
print("Standard Deviation:", df['TotalPay'].std())
```

```python
# Data Cleaning

# Since there are only a few missing values in the 'EmployeeName' column, we can drop rows with missing values.
df = df.dropna(subset=['EmployeeName'])

# For other columns with missing values, we can impute missing values with the mean or median of the corresponding column.
df['BasePay'] = df['BasePay'].fillna(df['BasePay'].mean())
df['OvertimePay'] = df['OvertimePay'].fillna(df['OvertimePay'].mean())
df['OtherPay'] = df['OtherPay'].fillna(df['OtherPay'].mean())
df['Benefits'] = df['Benefits'].fillna(df['Benefits'].mean())
```
